##### Задание 2: Сгенерировать OLAP-куб, где фиксируем продажи товаров в розничной сети по магазинам. Оси: 0: ID группы товаров, 1: ID магазина, 2: день месяца, в котором осуществляются продажи (предположим у нас в OLAP-кубе собраны продажи за месяц). Содержимое ячейки: продажи в тыс. рублей данной категории товаров в данном магазине в данный день.

```python
tovar_1 = (10 - (5))*np.random.random_sample((3, 29))+ 5 # Продажы первого товара в 3х магазинах за февраль 2020
#(составляют от 5 до 10 тысяч рублей в день)
tovar_1
```

```python
tovar_2 = (15 - (10))*np.random.random_sample((3, 29))+ 10 # Продажы второго товара в 3х магазинах за февраль 2020
#(составляют от 10 до 15 тысяч рублей в день)
tovar_2
```

```python
tovar_3 = (25 - (20))*np.random.random_sample((3, 29))+ 20 # Продажы 3-го товара в 3х магазинах за февраль 2020
#(составляют от 20 до 25 тысяч рублей в день)
tovar_3
```

```python
olap = np.vstack((tovar_1, tovar_2, tovar_3)) #объеденяем данные по товарам в один двумерный массив
olap.shape
```

```python
olap = np.reshape(olap, (3, 3, 29)) #объеденяем данные по товарам в OLAP-куб
olap.shape
```

```python
for tovar in 0,1,2:
        for shop in 0,1,2:
            (olap[tovar][shop][21]) *=1.4 
            (olap[tovar][shop][9]) *=1.4 
            (olap[tovar][shop][24]) *=1.4 
#Продажи выросли 10, 22 и 25 февраля на 40% перед праздником и в день получения зп
```

```python
olap
```

```python
np.mean(olap) #Cреднее значение продаж за 1 день во всех магазинах по всем товарам
```

```python
np.mean(olap[0]) #Cреднее значение продаж 1го товара
```

```python
np.mean(olap[2][0][1:10]) #Cреднее значение продаж 2го товара c 1 по 9 число
```

```python
np.mean(olap[2][0][19:28]) #Cреднее значение продаж 2го товара c 20 по 29 число
```

```python
np.min(olap[2]) #Минимальное значение продаж 3го товара
```

```python
np.sum(olap[1][2]) #Сумарное значение продаж 2го товара в 3 магазине
```

```python
np.sum(olap[1][2][21])/np.sum(olap[1][2][1]) #Разница продаж 2-го товара в 3 магазмне между 22 и 2 февраля
```
